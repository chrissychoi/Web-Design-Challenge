```python
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to Change These)
csv_file = "C:/Users/Christopher/Documents/GitHub/Web-Design-Challenge/WebVisualizations/Resources/cities.csv"

# Read Purchasing File and store into Pandas data frame
cities_data = pd.read_csv(csv_file)

cities_data = pd.DataFrame(cities_data)
cities_data = cities_data.rename(columns={"Max Temp":"Max_Temp","Wind Speed":"Wind_Speed"})

cities_data.head()
```

```python
plt.scatter(cities_data['Lat'],cities_data['Max_Temp'])
plt.title("Latitude v Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature")
plt.gcf().set_size_inches(12, 14)
plt.savefig("C:/Users/Christopher/Documents/GitHub/Web-Design-Challenge/WebVisualizations/assets/LatVsTemp.png")
plt.show()
```

```python
plt.scatter(cities_data['Lat'],cities_data['Cloudiness'])
plt.title("Latitude v Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.gcf().set_size_inches(12, 14)
plt.savefig("C:/Users/Christopher/Documents/GitHub/Web-Design-Challenge/WebVisualizations/assets/LatVsCloud.png")
plt.show()
```

```python
plt.scatter(cities_data['Lat'],cities_data['Wind_Speed'])
plt.title("Latitude v Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.gcf().set_size_inches(12, 14)
plt.savefig("C:/Users/Christopher/Documents/GitHub/Web-Design-Challenge/WebVisualizations/assets/LatVsWind.png")
plt.show()
```
